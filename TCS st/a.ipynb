```python
import matplotlib.pyplot as plt

# Data
data = [93.68, 91.66, 87.84]
labels = ['TCS', 'HDFC', 'CIPLA']

# Create bar chart
plt.bar(labels, data, color='#00008B', edgecolor='black', width = 0.3)

# Add titles and labels
plt.title('Accuracy')
plt.xlabel('Stocks')
plt.ylabel('Percentage')

# Set y-axis range from 0 to 100
plt.ylim(0, 100)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data = [93.68, 91.66, 87.84]
labels = ['TCS', 'HDFC', 'CIPLA']

# Create bar chart with custom bar width and color
plt.bar(labels, data, color='#00008B', edgecolor='black', width=0.3)

# Add titles and labels
plt.title('Stock Prices')
plt.xlabel('Stock Names')
plt.ylabel('Stock Prices')

# Set y-axis range from 0 to 100
plt.ylim(0, 100)

# Add grid lines at each percentage level
plt.grid(axis='y', which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```
